```python
import numpy as np
```

```python
audio_pred = np.random.rand(100, 7)
video_pred = np.random.rand(100, 7)
```

```python
audio_pred = audio_pred / audio_pred.sum(axis=1, keepdims=True)
video_pred = video_pred / video_pred.sum(axis=1, keepdims=True)
```

```python
alpha = 0.5

final_pred = alpha * audio_pred + (1 - alpha) * video_pred
```

```python
final_labels = np.argmax(final_pred, axis=1)

print(final_labels[:10])
```
